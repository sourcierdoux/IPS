```python
import pandas as pd
import numpy as np
import ast
```

```python
df=pd.read_csv('errors.csv')
true_1=(4,15)
true_2=(1.8,14)
true_3=(4,12.5)
true_4=(1.8,12)
true_5=(9,10)
FAN_POSITION=(7,14.5)
df=df.drop(columns=['5'],axis=1)
```

```python
df_2=pd.read_csv('errors_2.csv')
true_pos2=(8,10)
df_2=df_2.drop(columns=['2','3','4','5'],axis=1)
df_2=df_2.rename(columns={"1":"5"})
```

```python
df_3=pd.read_csv('errors_3.csv')
true_pos3=(4,7)
df_3=df_3.drop(columns=['1','3','4'],axis=1)
df_3=df_3.rename(columns={"2":"6"})
```

```python
df=pd.concat([df,df_2,df_3],axis=1)
```

```python
def convert_to_tuple(string):
    try:
        return ast.literal_eval(string)
    except (ValueError, SyntaxError):
        # Handle the error case or invalid format
        return None
```

```python

for column in df.columns:
    df[column] = df[column].apply(convert_to_tuple)
```

```python
true_values = [true_1, true_2, true_3, true_4, true_pos2, true_pos3]  # for 5 columns

# Initialize an empty dictionary to store errors
errors = {column: [] for column in df.columns}

def squared_diff(value, true_value):
    if value is None:
        return None
    return np.sqrt((value[0]*0.6 - true_value[0]*0.6)**2 + (value[1]*0.6 - true_value[1]*0.6)**2)

def angle_diff(value, true_value):
    if value is None:
        return None
    return np.rad2deg(np.abs(np.arctan2(FAN_POSITION[1]-true_value[1],FAN_POSITION[0]-true_value[0])-np.arctan2(FAN_POSITION[1]-value[1],FAN_POSITION[0]-value[0])))





# Convert errors dictionary to a DataFrame


# Display the errors DataFrame
```

```python
rmse={}
list_mse=np.array([])
# Calculate errors
for column, true_value in zip(df.columns, true_values):
        # Assuming each cell in the DataFrame is a tuple (x, y)
    mse = df[column].apply(lambda x: squared_diff(x, true_value))
    list_mse=np.append(list_mse,mse)
```

```python
list_angle=np.array([])
for column, true_value in zip(df.columns, true_values):
    angle = df[column].apply(lambda x: angle_diff(x, true_value))
    list_angle=np.append(list_angle, angle)
```

```python
average_positions = {}

for column in df.columns:
    sum_x, sum_y = 0, 0
    count = 0
    for value in df[column]:
        if value is not None:
            sum_x += value[0]
            sum_y += value[1]
            count += 1

    # Avoid division by zero
    if count > 0:
        average_x = sum_x / count
        average_y = sum_y / count
        average_positions[column] = (average_x, average_y)
    else:
        average_positions[column] = None
```
